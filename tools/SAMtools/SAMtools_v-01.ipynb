## samtools view

### SAM to BAM

```python
!curl https://s3.amazonaws.com/samtools-tutorial/sample.sam.gz > sample.sam.gz
!gzip -d sample.sam.gz
```

```python
# samtools view : file conversion
# -S : SAM input
# -b : BAM output
# output: sample.bam

!time samtools-1.16.1/samtools view -S -b sample.sam > sample.bam
```

```python
# samtools view
# head : top line data view

!samtools-1.16.1/samtools view sample.bam | head
```

```python
# samtools view
# head : top line data view

!samtools-1.16.1/samtools view sample.sam | head
```

### sort

```python
# samtools sort: sort in genome order
# output: sample.sorted.bam

!time samtools-1.16.1/samtools sort sample.bam -o sample.sorted.bam
```

```python
# view result

!samtools-1.16.1/samtools view sample.sorted.bam | head
```

### index

```python
# To extract alignments overlapping particular genomic regions. 
# output: sample.sorted.bam.bai

!time samtools-1.16.1/samtools index sample.sorted.bam
```

```python
!ls
```

```python
# To exploit the index to extract alignments from the 33rd megabase of chromosome 1.

!samtools-1.16.1/samtools view sample.sorted.bam 1:33000000-34000000
```

```python
# How many alignments are there in this region?

!samtools-1.16.1/samtools view sample.sorted.bam 1:33000000-34000000 | wc -l
```

```python
# inspecting the first five alignments in our BAM in detail.

!samtools-1.16.1/samtools view sample.sorted.bam | head -n 5
```

```python
# Count the total number of alignments.

!samtools-1.16.1/samtools view sample.sorted.bam | wc -l
```

```python
# Inspect the header.

!samtools-1.16.1/samtools view -H sample.sorted.bam
```

```python
# variants solely from paired-end sequences that aligned “properly” to the reference genome.
# -f  to report solely “proper pairs” 

!samtools-1.16.1/samtools view -f 0x2 sample.sorted.bam  | head -n 5
```

```python
# How many properly paired alignments are there?

!samtools-1.16.1/samtools view -f 0x2 sample.sorted.bam | wc -l
```

```python
# How many improperly paired alignments are there?
# 5026086 = 5003529 + 22557

!samtools-1.16.1/samtools view -F 0x2 sample.sorted.bam | wc -l
```
